```python
from gaussxw import gaussxw
from pylab import scatter,show,meshgrid,figure

N=30
a=0;b=1.0
x,w=gaussxw(N)
xp=0.5*(b-a)*x + 0.5*(b+a)
wp=0.5*(b-a)*w
xx,yy=meshgrid(xp,xp)
figure(figsize=(12,7))
scatter(xx,yy)
show()
```

```python
from gaussxw import gaussxw
from pylab import scatter,show,meshgrid,figure,fill_between
figure(figsize=(12,7))
fill_between([0.,1.],[0.,1.], color="g")

N=10
x,w=gaussxw(N)
xp=0.5*(b-a)*x + 0.5*(b+a)

for x1 in xp:
    yp=0.5*(x1-a)*x+0.5*(x1+a)
    scatter([x1]*len(xp),yp)
    
show()
```

```python
from gaussxw import gaussxw
from numpy import cos,pi
def f(x,y):
    return cos((pi/2)*(x-y))
a=-1;b=1
c=-1;d=1
N=10;M=10

x,w=gaussxw(N)
xp=0.5*(b-a)*x + 0.5*(b+a)
wp=0.5*(b-a)*w

y,v=gaussxw(N)
yp=0.5*(d-c)*y + 0.5*(c+d)
vp=0.5*(d-c)*v
s=0.0
for i in range(N):
    for j in range(M):
        s+=wp[i]*vp[j]*f(xp[i],yp[j])
        
print(s)
```

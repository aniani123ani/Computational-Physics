```python
def simpson(f,a,b,N):
    from numpy import sum,linspace
    S=0.0
    if N%2==1:
        raise ValueError("N must be an even integer")
    dx=(b-a)/N
    x=linspace(a,b,N+1)
    y=f(x)
    for i in range(1,N,2):
        S=S+4*y[i]
    for i in range(2,N,2):
        S=S+2*y[i]
    S=S+y[0]+y[N]
    S=S*(dx/3)
    return S

def f(x):
    return x*x
a=0.0;b=1.0;N=10
print(simpson(f,a,b,N))
print((1/3)*(b**3-a**3))
```

```python
def g(x):
    from numpy import sin
    return sin(x)
from pylab import linspace,plot,show,fill_between,pi
N=4;a=0;b=pi
# X and Y values are for plotting y=f(x)
X=linspace(a,b,100)
Y=g(X)
plot(X,Y,"k",lw=3)
# x and y values are for the trapezoid rule
x=linspace(a,b,N+1)
y=g(x)
print(x)
print(y)
for i in range(N):
    fill_between([x[i],x[i+1]],[y[i],y[i+1]])
```

```python
def simpsonp(f,a,b,N):
    from numpy import sum,linspace
    if N%2==1:
        raise ValueError("N must be an even integer")
    dx=(b-a)/N
    x=linspace(a,b,N+1)
    y=f(x)
    S=(dx/3)*(4.*sum(y[1::2])+2.*sum(y[2:-2:2])+y[0]+y[-1])
    return S

def f(x):
    return x*x
a=0.0;b=1.0;N=10
print(simpsonp(f,a,b,N))
print((1/3)*(b**3-a**3))
```

```python
from numpy import random,mean,std,exp,sqrt,pi
mu,sigma=0,0.1
s=random.normal(mu,sigma,10000)
print(abs(mu-mean(s)))
print(abs(sigma-std(s)))
```

```python
from pylab import plot,show
count,bins,ignored=hist(s,30,density=True)
plot(bins,1/(sigma*sqrt(2*pi))*exp(-(bins-mu)**2/(2*sigma**2)),linewidth=2,color='r')
show()
```

```python
from pylab import plot,show,xlabel,ylabel,figure
from numpy import random,sqrt,sum,pi,cos,sin,arange

n=10**4
x=random.rand(n)
y=random.rand(n)
r=sqrt(x**2+y**2)
inside=r<=1
outside=1-inside
print(n,sum(inside)/float(n),pi/4)

figure(figsize=(7,7))
phi=arange(0,pi/2,0.001)
plot(cos(phi),sin(phi),'k-',linewidth=2)
plot(x*inside,y*inside,'r.',markersize=1.0)
plot(x*outside,y*outside,'b.',markersize=1.0)
xlabel('$x$',fontsize=18)
ylabel('$y$',fontsize=18)
show()
```

```python
from pylab import plot,show,xlabel,ylabel,figure
from numpy import random
N=100
M=10
figure(figsize=[10,5])
steps=range(N)
for i in range(M):
    x=[0]*N
    for n in steps[1:]:
        x[n]=x[n-1]+random.choice([-1,+1])
    plot(steps,x,'o-')
xlabel("step")
ylabel("x(step)")
show()
```

```python
from numpy import sin
from random import random
def f(x):
    return(sin(1/(x*(2-x))))**2

N=100000
count=0
for i in range(N):
    x=2*random()
    y=random()
    if y<f(x):
        count+=1
I=2*count/N
print(I)
```

```python
from pylab import plot,show,fill_between,xlim,ylim,figure
from numpy import linspace,sin
def f(x):
    return(sin(1/(x*(2-x))))**2
x=linspace(0.000001,1.9999999,10000)
figure(figsize=(12,7))
xlim(0,2)
ylim(0,1)
plot(x,f(x),'b')
fill_between(x,f(x),facecolor='yellow')
show()
```

```python
from random import random
from numpy import arange
from pylab import plot,xlabel,ylabel,show

NTl=1000
NPb=0
tau=3.065*60
h=1.0
p=1-2**(-h/tau)
tmax=1000

tpoints=arange(0.0,tmax,h)
Tlpoints=[]
Pbpoints=[]

for t in tpoints:
    Tlpoints.append(NTl)
    Pbpoints.append(NPb)
    decay=0
    for i in range(NTl):
        if random()<p:
            decay+=1
    NTl-=decay
    NPb+=decay
    
plot(tpoints,Tlpoints)
plot(tpoints,Pbpoints)
xlabel("Time")
ylabel("Number of atoms")
show()
```

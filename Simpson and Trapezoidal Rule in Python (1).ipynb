```python
def trapezoid(f,a,b,N):
    h=(b-a)/N #N+1 points make N subintervals 
    s=0.5*f(a)+0.5*f(b)
    for k in range(1,N):
        s=s+f(a+k*h)
    return s*h
```

```python
def simpson2(f,a,b,N):
    from numpy import sum,linspace
    if N % 2==1:
        raise ValueError("N must be an even integer.")
        
    dx=(b-a)/N
    x=linspace(a,b,N+1)
    y=f(x)
    S=dx/3*(4.*sum(y[1::2])+2.*sum(y[2:-2:2])+y[0]+y[-1])
    return S
```
